###<center> **DATA PREPROCESSING**

```python
# Menghapus kolom yang tidak diperlukan
del data['Tanggal']
del data['Tahun']
```

```python
# Mendeteksi 8888 dan 9999 sebagai missing values
import numpy as np
data = data.replace({8888: np.nan, 9999: np.nan})
```

```python
# Melakukan label encoder pada kolom arah angin terbanyak
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Arah_Angin_Terbanyak'] = le.fit_transform(data['Arah_Angin_Terbanyak'])
```

```python
# Missing values
print(data.isnull().sum())
```

```python
# Mengatasi missing values
data = data.fillna(data.mean())
print(data.isnull().sum())
```

```python
# Mengonversi rentang curah hujan menjadi label kategori
bins = [0, 0.5, 20, 50, 100, 150, float('inf')]
labels = ['Berawan', 'Hujan Ringan', 'Hujan Sedang', 'Hujan Lebat', 'Hujan Sangat Lebat', 'Hujan Ekstrem']
data['Curah_Hujan'] = pd.cut(data['Curah_Hujan'], bins=bins, labels=labels, right=False)
data
```

```python
# Melakukan label encoding pada kolom curah hujan
data['Curah_Hujan'] = le.fit_transform(data['Curah_Hujan'])
```

###<center> **KLASIFIKASI MENGGUNAKAN ANN**

```python
X = data.drop(columns='Curah_Hujan') # Matriks fitur
y = data.Curah_Hujan # Vektor target

# Membagi data latih (90%) dan data uji (10%)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Random over sampling
from imblearn.over_sampling import RandomOverSampler
oversample = RandomOverSampler(sampling_strategy='auto')
X_ros, y_ros = oversample.fit_resample(X, y)

X_train, X_test, y_train, y_test = train_test_split(X_ros, y_ros, test_size=0.1, random_state=0)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Scaling
from sklearn.preprocessing import MinMaxScaler
columns = ['Suhu_Minimum','Suhu_Maksimum','Suhu_Rata2','Kelembapan_Rata2','Lama_Penyinaran_Matahari','Kecepatan_Angin_Maksimum','Arah_Angin_Saat_Kecepatan_Maksimum','Kecepatan_Angin_Rata2','Arah_Angin_Terbanyak']
scaler = MinMaxScaler()
X_train_scale = scaler.fit_transform(X_train[columns])
X_test_scale = scaler.transform(X_test[columns])
```

```python
# One hot encoding
from keras.utils import np_utils
y_train_encoded = np_utils.to_categorical(y_train, num_classes = 6)
y_test_encoded = np_utils.to_categorical(y_test, num_classes = 6)
```

```python
import numpy as np
from tensorflow.keras.callbacks import EarlyStopping
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers

# Model
def create_model(batch_size, learning_rate):
    model = Sequential()
    model.add(Dense(10,activation='relu', input_shape=(X_train.shape[1],)))
    model.add(Dense(32,activation='relu'))
    model.add(Dense(64,activation='relu'))
    model.add(Dense(32,activation='relu'))
    model.add(Dense(6,activation='sigmoid')) 
    model.compile(loss ='categorical_crossentropy',optimizer=optimizers.Adam(learning_rate=learning_rate),metrics=['accuracy'])
    early_stopping = EarlyStopping(monitor='val_accuracy', patience=20, restore_best_weights=True)
    history = model.fit(X_train_scale, y_train_encoded, epochs=100, batch_size=batch_size, validation_data=(X_test_scale, y_test_encoded), callbacks=[early_stopping])
    return model
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV

# Hypertuning
model_ann = KerasClassifier(build_fn=create_model)
params_ann= {'batch_size': [4, 8],
             'learning_rate':[0.001, 0.0001]}

ann = GridSearchCV(model_ann, param_grid=params_ann)
history1 = ann.fit(X_train_scale, y_train_encoded)
```

```python
# Parameter dan skor akurasi terbaik
print("Parameter Terbaik: ", history1.best_params_)
print("Skor Akurasi Terbaik: ", history1.best_score_)
```

```python
print("Best: %f using %s" % (history1.best_score_, history1.best_params_))
means = history1.cv_results_['mean_test_score']
stds = history1.cv_results_['std_test_score']
params = history1.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))

# Model terbaik
best_model = history1.best_estimator_.model
```

```python
# Membangun model kembali
model = Sequential()
model.add(Dense(10,activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(32,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(6,activation='sigmoid')) 

model.compile(loss ='categorical_crossentropy',optimizer=optimizers.Adam(learning_rate=0.001),metrics=['accuracy'])
model.summary()
```

```python
# Melatih data
early_stopping = EarlyStopping(monitor='val_accuracy', patience=20, restore_best_weights=True)
history2 = model.fit(X_train_scale, y_train_encoded, epochs=100, batch_size=4, validation_data=(X_test_scale, y_test_encoded), callbacks=[early_stopping])
```

```python
# Grafik loss
loss = history2.history['loss']
val_loss = history2.history['val_loss']

plt.plot(loss,'r', label = 'training loss')
plt.plot(val_loss,'b', label = 'validasi loss')
plt.legend()
```

```python
# Grafik akurasi
acc = history2.history['accuracy']
val_acc = history2.history['val_accuracy']

plt.plot(acc,'r', label = 'training accuracy')
plt.plot(val_acc,'b', label = 'validasi accuracy')
plt.legend()
```

```python
# Evaluasi
evaluation = model.evaluate(X_test_scale, y_test_encoded)
```

```python
# Prediksi
ann_pred = np.argmax(model.predict(X_test_scale),axis=1)
ann_pred
```

###<center> **CLUSTERING**

```python
# MENENTUKAN BANYAK CLUSTER MENGGUNAKAN SKOR SILHOUETTE
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Mencoba beberapa nilai jumlah klaster
for n_clusters in range(2, 10):
    # Membuat objek KMeans dengan jumlah klaster yang diinginkan
    kmeans = KMeans(n_clusters=n_clusters)
    
    # Melakukan clustering K-means
    labels = kmeans.fit_predict(data)
    
    # Menghitung skor silhouette
    silhouette_avg = silhouette_score(data, labels)
    
    # Menampilkan jumlah klaster dan skor silhouette rata-rata
    print("Jumlah Klaster =", n_clusters, "| Skor Silhouette =", silhouette_avg)
```

```python
# MENENTUKAN BANYAK CLUSTER MENGGUNAKAN METODE ELBOW
import matplotlib.pyplot as plt
# Menyimpan nilai inersia (inertia) untuk setiap jumlah klaster yang dicoba
inertias = []

# Mencoba beberapa nilai jumlah klaster
for n_clusters in range(1, 10):
    # Membuat objek KMeans dengan jumlah klaster yang diinginkan
    kmeans = KMeans(n_clusters=n_clusters)
    
    # Melakukan clustering K-means
    kmeans.fit(data)
    
    # Menyimpan nilai inersia dari model K-means
    inertias.append(kmeans.inertia_)
    
# Plot nilai inersia untuk setiap jumlah klaster
plt.plot(range(1, 10), inertias)
plt.xlabel('Jumlah Klaster')
plt.ylabel('Inersia')
plt.title('Metode Elbow')
plt.show()
```

```python
# MODEL CLUSTERING DENGAN BANYAK CLUSTER 2
# banyaknya cluster diambil dari skor silhouette tertinggi
# banyaknya cluster juga dapat dilihat pada siku plot elbow
clustering = KMeans(n_clusters = 2, max_iter = 300) 
y_kmeans = clustering.fit(data)
```

```python
# MEMBERI LABEL KOLOM HASIL CLUSTER
data['KMeans_Clusters'] = y_kmeans.labels_
data
```

```python
# EVALUASI MODEL
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

# Menghitung Silhouette Coefficient
silhouette_avg = silhouette_score(data, y_kmeans)
print("Silhouette Coefficient:", silhouette_avg)

# Menghitung Davies-Bouldin Index
db_index = davies_bouldin_score(data, y_kmeans)
print("Davies-Bouldin Index:", db_index)

# Menghitung Calinski-Harabasz Index
ch_index = calinski_harabasz_score(data, y_kmeans)
print("Calinski-Harabasz Index:", ch_index)
```
